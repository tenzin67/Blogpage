MATHS OPERATION IN TENSOR

```python
a = torch.tensor([1.,2.,3.])
b = torch.tensor([4.,5.,6.])
# Tensor Addition
torch.add(a,b)
a+b

#Tensor Subtraction
torch.sub(a,b)
a-b

#Tensor Multiplication
torch.mul(a,b)
a*b

#Tensor Division
torch.div(a,b)
a/b
```
